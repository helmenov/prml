Plot loss functions
==========================
http://scikit-learn.org/stable/auto_examples/linear_model/plot_sgd_loss_functions.html

```python
import numpy as np

def LogisticLoss(y,g):
    return np.log(1 + np.exp(-y*g))

def HingeLoss(y, g):
    return np.maximum(1.0 - y*g, 0)

def SquaredLoss(y, g):
    return 0.5 * (y - g)**2
```

```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

xmin, xmax = -3, 3
y = 1
gx = np.linspace(xmin, xmax, 100)
lw = 2
ax = plt.axes()
plt.plot([xmin, 0, 0, xmax], [1, 1, 0, 0], color='gold', lw=lw, label="Zero-one loss")
plt.plot(gx, LogisticLoss(y, gx), color='cornflowerblue', lw=lw, label="Logistic loss")
plt.plot(gx, HingeLoss(y, gx), color='red', lw=lw, label="Hinge loss")
plt.plot(gx, SquaredLoss(y, gx), color='gray', ls='--', lw=lw/2, label="Squared loss")

plt.xlim(xmin, xmax)
plt.ylim(0, 4)
plt.grid(True, ls='--', lw=0.2)
plt.legend(loc="upper right", fontsize=16)
plt.xlabel("$g(x)$", fontsize=20)
plt.ylabel("loss$(y=1, g(x))$", fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.set_aspect('equal')
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
plt.tight_layout()

#plt.savefig('loss_functions.png', transparent=True)
plt.show()
```
